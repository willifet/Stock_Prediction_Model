```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import tensorflow
```

```python
start = '2010-01-01'
end = '2025-04-01'
stock = 'TSLA'

data = yf.download(stock, start, end)
```

```python
#replacing date as an index
data
```

```python
#creating a moving average chart
ma_100_days = data.Close.rolling(100).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r')
plt.plot(data.Close, 'g')
plt.show()
```

```python
ma_200_days = data.Close.rolling(200).mean()
```

```python
plt.figure(figsize=(8,6))
plt.plot(ma_100_days, 'r')
plt.plot(ma_200_days, 'b')
plt.plot(data.Close, 'g')
plt.show()
```

```python
data.dropna(inplace=True)
```

```python
#Splitting data for training and testing (80% Train and 20% Test)

data_train = pd.DataFrame(data.Close[0: int(len(data)*0.80)])
data_test = pd.DataFrame(data.Close[int(len(data)*0.80): len(data)])
```

```python
data_train.shape[0], data_test.shape[0]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
#Scaling the training dataset btw 0 & 1

data_train_scale = scaler.fit_transform(data_train)
```

```python
#Taking first 100days data to calculate the next 100 days 

x = []
y = []

for i in range(100, data_train_scale.shape[0]):
    x.append(data_train_scale[i-100:i])
    y.append(data_train_scale[i,0])

x, y = np.array(x), np.array(y)
```

```python
#Creating a Model
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
#Model building

model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True,
               input_shape = ((x.shape[1],1))))
model.add(Dropout(0.2))

model.add(LSTM(units = 60, activation= 'relu', return_sequences = True))
model.add(Dropout(0.3))

model.add(LSTM(units = 80, activation = 'relu', return_sequences =True))
model.add(Dropout(0.4))

model.add(LSTM(units = 120, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(units =1, activation= 'relu'))

```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x,y, epochs= 50, batch_size =32, verbose = 1)
```

```python
model.summary()
```

```python
pas_100_days = data_train.tail(100)
```

```python
data_test = pd.concat([pas_100_days, data_test], ignore_index=True)
```

```python
data_test_scale = scaler.fit_transform(data_test)
```

```python
#Testing the model

x = []
y = []

for i in range(100, data_test_scale.shape[0]):
    x.append(data_test_scale[i-100:i])
    y.append(data_test_scale[i,0])

x, y = np.array(x), np.array(y)
```

```python
y_predict = model.predict(x)
```

```python
1/scaler.scale_
```

```python
plt.figure(figsize=(10,8))
plt.plot(y_predict, 'r', label = 'Predicted price')
plt.plot(y, 'g', label = 'Original Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
model.save('Stock Prediction Model.keras')
```
